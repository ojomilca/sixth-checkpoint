```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/jessi/Downloads/iris.data.txt", names=['sepal length', 'sepal width', 'petal length', 'petal width','species'])
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test= train_test_split(df,test_size=0.28)
df_train
```

```python
def distance(V1,V2):
    S=0 
    for f in ['sepal length', 'sepal width', 'petal length', 'petal width']:
        S=S+(V1.get(f, 0)-V2.get(f, 0))**2
    return np.sqrt(S)
```

```python
import operator

def getNeighbors(trainingSet, testInstance, k):

    distances = []

    length = len(testInstance)-1

    for x in range(len(trainingSet)):

        dist = distance(testInstance, trainingSet[x])

        distances.append((trainingSet[x], dist))

        distances.sort(key=operator.itemgetter(1))

    neighbors = []

    for x in range(k):

        neighbors.append(distances[x][0])

    return neighbors
```

```python
trainSet = df_train.to_dict(orient="records")

testInstance = {'sepal length': 5,
  'sepal width': 5,
  'petal length': 5}

k = 10

neighbors = getNeighbors(trainSet, testInstance, k)
```

```python
def getResponse(neighbors):

    classVotes = {}

    for x in range(len(neighbors)):

        response = neighbors[x]['species']

        if response in classVotes:

            classVotes[response]=classVotes[response]+1
        else:
            classVotes[response]=1

    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)

    return sortedVotes[0][0]

#We can test out this function with some test neighbors, as follows:


response = getResponse(neighbors)

print(response)
```

```python
def getAccuracy(testSet, predictions):
    correct=0
    for instance, pred in zip(testSet,predictions):
        if instance['species']==pred:
            correct=correct+1
    return (correct/float(len(testSet))) * 100.0

#We can test this function with a test dataset and predictions, as follows:

testSet = df_test.to_dict(orient="records")

predictions = [getResponse(getNeighbors(trainSet, x, k)) for x in testSet]

accuracy = getAccuracy(testSet, predictions)

print(accuracy)
```
